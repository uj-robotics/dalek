```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import pickle
import matplotlib.pylab as plt
import sklearn
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
import numpy as np
from sklearn.datasets import fetch_lfw_people, fetch_mldata
%matplotlib inline
```

```python
# Load
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

X, Y = lfw_people.data.astype("float64"), lfw_people.target
X = preprocessing.scale(X)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=1)
N = 1288
model = LogisticRegression()
model.fit(X_train[0:N], Y_train[0:N])

Y_test_predicted = model.predict(X_test)
print "Dokładność modelu wytrenowanego na ",N, " przykladach to: ",100*sklearn.metrics.accuracy_score(Y_test, Y_test_predicted), "%"
```
